# Data Preparation

```python
#Checking the null values
Bike.isnull().sum()
```

```python
##Drop the null values
Bike.drop(['instant'],axis=1,inplace=True)
Bike.drop(['dteday'],axis=1,inplace=True)
Bike.drop(['casual','registered'],axis=1,inplace=True)
```

```python
Bike.head()
```

```python
#In given dataset, we can say season,yr,mnth,holiday,weekday,workingday,weathersit  are categorical variables, so replace them with meaningful name
Bike['season'].replace({1:"spring",2:"summer",3:"fall",4:"winter"},inplace=True)
Bike['weathersit'].replace({1:"Clear_Few Clouds",2:"Mist_cloudy",3:"Light rain_Light snow_Thunderstorm",4:'Heavy Rain_Ice Pallets_Thunderstorm_Mist'},inplace=True)
Bike['weekday'].replace({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thursday",5:"Friday",6:"Saturday"},inplace=True)
```

# Dummy Variables

```python
# convert categorical variable into dummy/indicator variables 
SeasonCondition=pd.get_dummies(Bike['season'],drop_first=True)
WeatherCondition=pd.get_dummies(Bike['weathersit'],drop_first=True)
DayOfWeek=pd.get_dummies(Bike['weekday'],drop_first=True)
MonthName=pd.get_dummies(Bike['mnth'],drop_first=True)
```

```python
SeasonCondition
```

```python
WeatherCondition
```

```python
DayOfWeek
```

```python
MonthName
```

```python
Bike=pd.concat([Bike,SeasonCondition],axis=1)
Bike=pd.concat([Bike,WeatherCondition],axis=1)
Bike=pd.concat([Bike,DayOfWeek],axis=1)
Bike=pd.concat([Bike,MonthName],axis=1)
```

```python
Bike.head()
```

```python
#Droing the older variable
Bike.drop(['season'],axis=1,inplace=True)
Bike.drop(['weathersit'],axis=1,inplace=True)
Bike.drop(['weekday'],axis=1,inplace=True)
Bike.drop(['mnth'],axis=1,inplace=True)
Bike.head()
```

# Splitting the Data into Training and Testing Sets


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
np.random.seed(0)
Bike_train,Bike_test = train_test_split(Bike, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
Bike_train.head()
```

```python
Bike_test.head()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['atemp', 'hum', 'windspeed', 'cnt']
Bike_train[num_vars] = scaler.fit_transform(Bike_train[num_vars])
Bike_train.head()
```

# Dividing into X and Y sets for the model building

```python
y_train = Bike_train.pop('cnt')
X_train = Bike_train
```

```python
X_train.head()
```

```python
y_train.head()
```

# Building our model
This time, we will be using the LinearRegression function from SciKit Learn for its compatibility with RFE (which is a utility from sklearn)

# RFE
Recursive feature elimination

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# RFE method for feature selection, and lets take 15 variable

lm = LinearRegression()
lm.fit(X_train, y_train)
rfe = RFE(lm, n_features_to_select=15)           
rfe = rfe.fit(X_train, y_train)

```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
X_train_rfe
```
